# Regression Analysis

```python
import numpy as np
```

```python
x = np.random.rand(20) * 8 - 4 # 把資料控制在 -4 到 +4
x
```

```python
y = np.sin(x) + np.random.rand(20) * 0.2
y
```

```python
w = np.polyfit(x, y, 1)
w
```

```python
import matplotlib.pyplot as plt
def polyfit(w):
    f = np.poly1d(w)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('using plotfitfunction')
    plt.grid()
    plt.scatter(x, y, marker='x', c='red')
    xx = np.linspace(-4, 4, 100)
    plt.plot(xx, f(xx), color='green')
    plt.show()
```

```python
polyfit(w)
```

```python
w = np.polyfit(x, y, 2)
w
```

```python
polyfit(w)
```

```python
w = np.polyfit(x, y, 3)
w
```

```python
polyfit(w)
```
